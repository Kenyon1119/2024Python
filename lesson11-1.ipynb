```python
import pyinputplus as pyip #猴子使用者亂輸入
import random   #內建套件
nums=pyip.inputInt(f"請輸入學生數量:")
students:list[dict]=[]  #List
for _ in range(nums): #迴圈圈數指定
    student:dict[str,int]={}    #建置list的dict
    for subject in ['國文','英文','數學']:
        student[subject]=random.randint(50,100) 
    students.append(student)
print(f'{nums}')
students
```

```python
import pyinputplus as pyip #猴子使用者亂輸入
import random   #內建套件
from pprint import pprint   #呼叫排序輸出pprint
#---------------------------------------
def get_students(n:int)->list[dict]:
    for _ in range(n): #迴圈 0-49，共50圈
        student:dict[str,int]={}    #形容並建值50組
        for subject in ['國文','英文','數學']:
            student[subject]=random.randint(50,100) 
        students.append(student)
    return students
#------------方程式-簡化邏輯-------------
nums=pyip.inputInt(f"請輸入學生數量:")
stus=get_students(n=nums)
pprint(stus)
#csv    純文字檔，儲存表格資料
#excel  非文字檔，儲存表格資料
#json   純文字檔，複雜的資料結構
```
