```python
##建立學生分數
from random import randint
stu1:dict[str,int]={
    '國文':randint(50,100),
    '英文':randint(50,100),
    '數學':randint(50,100),
    '社會':randint(50,100),
    '自然':randint(50,100),
    }
stu1
```

```python
stu2:dict[str,int]={} ##建立空的dictionary
for key in ['國文','英文','數學','社會','自然']:
    stu2[key]=randint(50,100)
stu2
```

```python
##dict的comprehension -> {key:value for...in}

stu3:dict[str,int]={key:randint(50,100)for key in ['國文','英文','數學','社會','自然']}
stu3
```

```python
##建立50個學生各5個分數
from pprint import pprint
students:list[dict[str,int]]=[]
for _ in range(50):
    stu:dict[str,int]={key:randint(50,100)for key in ['國文','英文','數學','社會','自然']}
    students.append(stu)
pprint(students)
```

```python
for student in students:
    print(student)
```

```python
for student in students:
    for key in ['國文','英文','數學']:
        print(f'{key}={student[key]}')
    print('=======')
```
